```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df.ffill( inplace =True)
df.bfill(inplace = True)
```

```python
sns.boxplot(x="Sex", y="Age" , data = df, hue = "Survived")
# plt.legend(title='Survived', labels=['No', 'Yes'])
plt.title("hii")
```

Observations:
The box plot shows the distribution of ages for male and female passengers, categorized by their survival status (survived or not)##### . For both genders, the median age of passengers who survived appears to be slightly lower than those who did not surviv##### e. There is more variability in the age distribution of male passengers compared to female passengers, especially among those who did not survi##### ve. Female passengers generally have a higher survival rate compared to male passengers across different age gro##### ups. There are outliers in both male and female groups, indicating certain age groups where the survival rate deviates significantly from the ### norm. These observations provide insights into the relationship between age, gender, and survival status among Titanic pass.
